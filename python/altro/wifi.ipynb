```python
import cv2
import numpy as np
import socket
import subprocess

height = 240
width = 320
NUMERO_BYTE_PER_PIXEL = 2

ssid = "martin_k"
password = "perselacappa"

subprocess.run(["nmcli", "dev", "wifi", "connect", ssid, "password", password])
```

```python

host = "192.168.3.1"   # indirizzo del server
port = 80             # porta del server

# crea un socket TCP
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# connessione
s.connect((host, port))
print("Connesso!")
```

```python

buff_size = height*width*NUMERO_BYTE_PER_PIXEL

data_raw = b""
while len(data_raw) < buff_size:
    chunk = s.recv(buff_size - len(data_raw))
    print(len(chunk))
    if not chunk:
        raise RuntimeError("Connessione chiusa prima di ricevere tutti i dati")
    data_raw += chunk

print("Ricevuti:", len(data_raw), "byte")


print(f"Dimensioni previste: {height*width*NUMERO_BYTE_PER_PIXEL}")

data = np.frombuffer(data_raw, dtype=np.uint8)


img = data.reshape((height, width , NUMERO_BYTE_PER_PIXEL))

# img = img[:, :, ::-1]  
# img = cv2.cvtColor(img, cv2.COLOR_BGR5652BGR)

cv2.imwrite("immagine.png" ,img)
```
